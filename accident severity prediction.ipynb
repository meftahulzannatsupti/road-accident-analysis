### 6. Identify accidents in our region: Kingston upon Hull, Humberside, and the East Riding of Yorkshire etc.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

#Filter rows containing specific region names in the 'lsoa01nmw' column
region_keywords = ["Kingston upon Hull", "North Lincolnshire", "North East Lincolnshire", "East Riding of Yorkshire"]
final_df['region'] = np.where(final_df['lsoa01nmw'].str.contains('|'.join(region_keywords), na=False), final_df['lsoa01nmw'], 'Not Included')

#Rename variations of region names to their standard names
final_df['region'] = final_df['region'].replace({"Kingston upon Hull.*": "Kingston upon Hull", 
                                                 "North Lincolnshire.*": "North Lincolnshire",
                                                 "North East Lincolnshire.*": "North East Lincolnshire",
                                                 "East Riding of Yorkshire.*": "East Riding of Yorkshire"}, regex=True)

#Extract the latitude and longitude for clustering
region_df = final_df[final_df['region'] != 'Not Included']
coordinates = region_df[['latitude', 'longitude']]

#Combine latitude and longitude into a 2D array
input_array = coordinates.values

#Define colors for each region
colors = {'Kingston upon Hull': 'pink', 'North Lincolnshire': 'salmon', 'North East Lincolnshire': 'skyblue', 'East Riding of Yorkshire': 'blue'}

#Calculate the sum of squared distances for different number of clusters
ssd = []
K = range(1, 11)  # Check from 1 to 10 clusters
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(input_array)
    ssd.append(kmeans.inertia_)

#Determine the optimal number of clusters from the Elbow Method plot
diff = np.diff(ssd)
diff_r = diff[1:] / diff[:-1]
optimal_clusters = K[np.argmin(diff_r) + 1]  # Adding 1 because of zero-based indexing
print("Optimal number of clusters:", optimal_clusters)

#Perform KMeans clustering with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
region_df['cluster'] = kmeans.fit_predict(input_array)

#Plot the clusters with different colors for different regions
plt.figure(figsize=(15, 10))
for region_name, group in region_df.groupby('region'):
    plt.scatter(group['longitude'], group['latitude'], label=region_name, color=colors[region_name])

#Plot the cluster centers as stars
for center in kmeans.cluster_centers_:
    plt.scatter(center[1], center[0], marker='*', s=200, color='black')

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Clustering of Accidents in Kingston upon Hull, North Lincolnshire, North East Lincolnshire, and East Riding of Yorkshire')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report


features_list = ['hour', 'age_of_driver', 'sex_of_driver', 'speed_limit', 
                 'weather_conditions', 'light_conditions', 'vehicle_type', 
                 'age_of_casualty', 'age_of_vehicle']
target = 'accident_severity'


#Balancing the target variable
final_df[target] = final_df[target].apply(lambda x: 1 if x == 1 else 0)

#Remove rows with missing values
final_df = final_df[features_list + [target]].dropna()

#Normalize the Data
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(final_df.drop(columns=[target]))
scaled_df = pd.DataFrame(scaled_features, columns=features_list).join(final_df[target])

#Split the Data
X = scaled_df.drop(columns=[target])
y = scaled_df[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#Train the Random Forest Model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

#Evaluate the Model
y_pred = rf_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)


print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("\nClassification Report:\n", class_report)

# Visualization: Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False, square=True)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report

#Define features and target variable
features_list = ['hour', 'age_of_driver', 'sex_of_driver', 'speed_limit', 
                 'weather_conditions', 'light_conditions', 'vehicle_type', 
                 'age_of_casualty', 'age_of_vehicle']
target = 'accident_severity'

#Balance the target variable
final_df[target] = final_df[target].apply(lambda x: 1 if x == 1 else 0)

#Remove rows with missing values
final_df = final_df[features_list + [target]].dropna()

#Normalize the Data
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(final_df.drop(columns=[target]))
scaled_df = pd.DataFrame(scaled_features, columns=features_list).join(final_df[target])

#Split the Data
X = scaled_df.drop(columns=[target])
y = scaled_df[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#Train the Decision Tree Model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

#Evaluate the Model
y_pred = dt_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("\nClassification Report:\n", class_report)

#Visualization: Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False, square=True)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```
